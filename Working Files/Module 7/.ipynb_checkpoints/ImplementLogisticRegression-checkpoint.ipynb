## Importing dataset

```python
import pandas as pd
data=pd.read_csv("C:/Users/Ganapathi surya.DESKTOP-JJB2T5O/Machine Learning Internshala/Module Files/Module 7/churn_prediction_simple.csv")
```

```python
data.head()
```

## Seperating dependent and independent variables

```python
x=data.drop(columns=["churn","customer_id"])
```

```python
y=data["churn"]
```

## Scaling the dataset

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
scaled_x=scaler.fit_transform(x)
```

## Splitting the dataset

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(scaled_x,y,test_size=0.8)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Model Building

```python
from sklearn.linear_model import LogisticRegression as LR
classifier=LR(class_weight='balanced')
classifier.fit(x_train,y_train)
```

```python
pred=classifier.predict(x_test)
```

```python
pred
```

```python
pred_prob=classifier.predict_proba(x_test)
```

```python
pred_prob
```

## Accuracy

```python
classifier.score(x_test,y_test)
```

## Precision

```python
from sklearn.metrics  import precision_score
```

```python
prec_score=precision_score(y_test,pred)
```

```python
prec_score
```

## Recall

```python
from sklearn.metrics  import recall_score
```

```python
rec_score=recall_score(y_test,pred)
```

```python
rec_score
```

## F1 score

```python
f1=2/((1/prec_score)+(1/rec_score))
```

```python
f1
```

```python
from sklearn.metrics import f1_score
```

```python
f=f1_score(y_test,pred)
```
