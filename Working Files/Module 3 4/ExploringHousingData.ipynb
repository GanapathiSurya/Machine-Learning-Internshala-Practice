### Imputing on Independent variables 

1) imputing- Replacing values with their column's median, mean or mode.
2) Imputing is generally preferred for Independent variables.
3) For Numerical columns - Imputing strategy is mean or median.
4) For Non numeric columns(Object variables) - Imputing strategy is mode.

```python
Raw_housing["No of Bathrooms"]
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='median')
Raw_housing["No of Bathrooms"] = imputer.fit_transform(Raw_housing["No of Bathrooms"])
```
