```python
sale_price=transformed_housing["Sale_Price"].head(30)
flat_area=transformed_housing["Flat Area (in Sqft)"].head(30)
```

```python
sale_price.count()
```

```python
flat_area.count()
```

```python
sample_data=pd.DataFrame({'sale_price':sale_price,'flat_area':flat_area})
```

```python
sample_data
```

```python

plt.figure(dpi=100)
plt.scatter(sample_data.flat_area,sample_data.sale_price,color="violet")
plt.xlabel("Flat Area")
plt.ylabel("Sale Price")
plt.title("Sale Price vs Flat Area")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error as mse
c=0
m=0
line=[]
for i in range(len(sample_data)):
    line.append(sample_data.flat_area[i]*m+c)
plt.scatter(sample_data.flat_area,sample_data.sale_price)
plt.plot(sample_data.flat_area,line,label="m=0 & c=0")
plt.xlabel("Flat Area")
plt.ylabel("Sale Price")
plt.legend()
mse=mse(sample_data.sale_price,line)
plt.title("Slope "+str(m)+"with MSE "+str(mse))
plt.show()
```

 ## Drawing Cost Function Curve w.r.to random slopes

```python
from sklearn.metrics import mean_squared_error as mse
def slope_error(slope,intercept):
    sale=[]
    for j in range(len(sample_data)):
        t=sample_data.flat_area[j]*slope+intercept
        sale.append(t)
    MSE = mse(sample_data.sale_price,sale)
    return MSE
```

```python
slope=[i/10 for i in range(5000)]
costs=[]
for j in slope :
    costs.append(slope_error(j,35732))
```

```python
cost_table=pd.DataFrame({'slope':slope,'cost':costs})
```

```python
cost_table
```

```python
plt.plot(cost_table.slope,cost_table.cost, label="Cost for each slope")
plt.xlabel("Slope")
plt.ylabel("Cost")
plt.legend()
```

```python
least_cost_slope=cost_table["slope"][cost_table["cost"]==cost_table["cost"].min()]
print(least_cost_slope)
```

```python
cost_table.iloc[2236]
```

 ## Drawing Cost Function Curve w.r.to random intercepts

```python
def intercept_error(slope,intercept):
    sale=[]
    for j in range(len(sample_data)):
        sale.append(sample_data.flat_area[j]*slope+intercept)
    return mse(sample_data.sale_price,sale)
        
```

```python
intercepts=[i for i in range(5000,50000)]
costi=[]
for j in intercepts:
    costi.append(intercept_error(219.3,j))
    
```

```python
costi_table=pd.DataFrame({'intercept':intercepts,'cost':costi})
```

```python
costi_table
```

```python
plt.plot(costi_table.intercept,costi_table.cost,label="Cost for each intercept")
```

```python
least_cost_intercept=costi_table["intercept"][costi_table["cost"]==costi_table["cost"].min()]
print(least_cost_intercept)
```
