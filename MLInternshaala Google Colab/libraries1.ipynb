```
import numpy as np
Raw_Housing_Data['Ever Renovated']=np.where(Raw_Housing_Data['Renovated Year']==0,'No','Yes') 
```

```
Raw_Housing_Data['Purchase_year']=pd.DatetimeIndex(Raw_Housing_Data['Date House was Sold']).year
```
