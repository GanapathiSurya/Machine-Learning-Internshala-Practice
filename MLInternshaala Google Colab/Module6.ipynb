```
numerical_data=data.drop(columns=['ID','Date House was Sold','Sale Price','Waterfront View','No of Times Visited','Condition of the House','Overall Grade','Zipcode'])
numerical_data.info()
```

```
from sklearn.preprocessing import Normalizer
norm=Normalizer()
norm_data = norm.fit_transform(numerical_data)
norm_data =pd.DataFrame(norm_data,columns=numerical_data.columns)
norm_data.head()
```

```
tmp=norm_data.var()
tmp
```

```
T=1e-04
for i in tmp:
  if i < T:
    norm_data.drop(columns=[tmp[tmp==i].index[0]],inplace = True)
norm_data.var()
```
