```
Raw_Housing_Datann['Sale Price'].describe()
```

```
def limit_imputer(value):
  if value>upper_limit:
    return upper_limit
  elif value <lower_limit:
    return lower_limit
  else:
    return value
```

```
Raw_Housing_Datann['Sale Price'].describe()
```

```
Raw_Housing_Datann['Sale Price']=Raw_Housing_Datann['Sale Price'].apply(limit_imputer)
```

```
Raw_Housing_Datann['Sale Price'].describe()
```

```
Raw_Housing_Datann['Sale Price'].describe()
```

```
Raw_Housing_Datann.dropna(inplace=True, axis=0,subset=['Sale Price'])
```

```
numerical_columns=['No of Bathrooms','Flat Area (in Sqft)','Area of the House from Basement (in Sqft)','Latitude','Longitude','Living Area after Renovation (in Sqft)']
```

```
import numpy as np
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='median')
Raw_Housing_Datann[numerical_columns]=imputer.fit_transform(Raw_Housing_Datann[numerical_columns])
```

```
imputer=SimpleImputer(missing_values=np.nan,strategy='median')
Raw_Housing_Datann['Zipcode']=imputer.fit_transform(Raw_Housing_Datann['Zipcode'])
```
